```python
# libraries used
import os
import glob
import statistics
import csv
import pandas as pd
import numpy as np
import scipy.stats as stats

```

```python

# use glob to get all the csv files 
# in the folder
# specifying the path to csv files
path = "../data/AReM/*"
  
# csv files in the path
csv_files = glob.glob(path + "/*.csv")
  
# loop over the list of csv files
for f in csv_files:
   # read the csv file
    data= pd.read_csv(f)
    data = data.drop(columns={data.columns[0]: 'time'}) #rename column test
    #data = data.drop('time', axis=1)# drop column test
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
    # print the content
    print('Content:')
    #display(data)
    print(data)
    
         
```

```python

stats = np.empty((0, 42))
for i in data:
    minimum = data.min()
    maximum = data.max()
    mean = data.mean()
    median = data.median()
    standarddev = data.std()
    firstquartile = data.quantile(q = .25)
    thirdquartile = data.quantile(q = .75)
    data_temp = np.array([minimum, maximum, mean, median, standarddev,firstquartile,thirdquartile]).reshape((1,42))
    stats = np.vstack((stats, data_temp))
stats = stats[1:]
data2 = pd.DataFrame(stats)
data2.columns=["min1", "max1", "mean1", "median1", "standarddev1", "1stQuart1", "3rdQuart1", "min2", "max2", "mean2", "median2", "standarddev2", "1stQuart2", "3rdQuart2", "min3", "max3", "mean3", "median3", "standarddev3", "1stQuart3", "3rdQuart3", "min4", "max4", "mean4", "median4", "standarddev4", "1stQuart4", "3rdQuart4", "min5", "max5", "mean5", "median5", "standarddev5", "1stQuart5", "3rdQuart5", "min6", "max6", "mean6", "median6", "standarddev6", "1stQuart6", "3rdQuart6"]
```
